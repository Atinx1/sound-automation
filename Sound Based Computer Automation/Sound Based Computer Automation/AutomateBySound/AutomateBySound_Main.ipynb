```python
########################################################
### This is a python file(.ipynb) for Jupyter      #####
### The code is property of EFY magazine            ####
### https://efymag.com                 #################

### Author: Vijaykumar Sajjanar                 ########
### email: ec.vijaykumar@bldeacet.ac.in             ####
### Phone: (+91)9980 67 7175                 ###########
########################################################


# Automate Slideshow/Music/Video SCRIPT
# Requires 5 packages, time-sounddevice-matplotlib-numpy-pyautogui
# matplotlib-numpy-time come preinstalled with Anaconda package

                                                         # Prompt user to select work
print("Welcome, What would you like to automate today?")
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("Press Number Followed by Enter")
print("SlideShow-1, Music-2, Video-3")
                                                        # Prompt user to enter integer input
selection=int(input());
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("For how many minutes the automation should run?")
                                                        # Prompt user to enter integer input
runtime=int(input());
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
                                                        # Exit if wrong selection
if (selection>3 or selection<1):
    print("Invalid Response.Rerun the script")
elif (selection==1):
    print("Your SlideShow will be Automated in 5 seconds.")
    print("It will run for",runtime,"minutes")
    print("You can change Slide by snap/clap. Good luck!")
elif (selection==2):
    print("Your Music will be Automated in 5 seconds.")
    print("It will run for",runtime,"minutes")
    print("You can mute/unmute Music by snap/clap. Good luck!")
elif (selection==3):
    print("Your Video will be Automated in 5 seconds.")
    print("It will run for",runtime,"minutes")
    print("You can pause/play Video by snap/clap. Good luck!")
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
                                                        # Wait 5 seconds for user to get ready
import time
time.sleep(5)

                                                        # Find range of for loop by runtime
                                                        # (User input minutes)*60 seconds/5
                                                        # 5 is recording seconds
runtimevar=runtime*12;
for i in range (0,runtimevar): 
    import sounddevice as sd                            # Sounddevice is to be installed using pip
    RATE = 44100                                        # Sample rate
    RECORD_SECONDS = 5                                  # Duration of recording
                                                        # Record from default input sound device
    myrecording = sd.rec(int(RECORD_SECONDS * RATE), samplerate=RATE, channels=2)
    sd.wait() 
                                                        # Analyse np array and find highest value
    import numpy as np
    data=np.amax(myrecording)
                                                        # Import pyautogui to automate keyboard keys
    import pyautogui
                                                        # Edit '0.1' based on microphone sensitivity
    if (data > 0.1):
        if(selection==1):
            pyautogui.press('right')
        if(selection==2):
            pyautogui.press('volumemute')
        if(selection==3):
            pyautogui.press('space',interval=0.25)
                    
            
########################################################
################## END OF SCRIPT #######################
########################################################
```
